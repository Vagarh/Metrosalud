# Caracterizacion de la poblacion del Programa Buen Comienzo Modalidad Familiar Febrero 2023

### Objetivos
### General
Identificar las características y necesidades de los participantes del programa Buen Comienzo Modalidad Familiar , de manera tal que permita contar con información precisa que pueda ser requerida en la UEABC y la E.S.E METROSALUD para adecuar la oferta institucional, diseñar o implementar atenciones que impacten positiamente en la poblacion de la ciudad 

Este producto comprende el soporte para el** cumplimiento de la estrategia Nº2 del contrato** “*Adelantar procesos de monitoreo de indicadores de primera infancia (en salud) del Municipio, con el fin de adelantar acciones conjuntas para la promoción y garantía de derechos, a través de la gestión territorial. “*

Se toma como referencia "[Guía Metodológica Sistema Nacional de Seguimiento a la Primera Infancia, Infancia y Adolescencia en Colombia](https://https://www.minsalud.gov.co/sites/rid/Lists/BibliotecaDigital/RIDE/VS/ED/GCFI/INFAN_00116.12.2013.pdf) considerando las particularidades de la modalidad de la atencion 

La fuente de informacion comprede



*   Base de Datos Buen Comienzo - Corte 28/Febrero/2023( emitido por la UEABC)
*   Registro Mensual de Tamizajes Nutricionales-Febrero-componente de nutricion Metrosalud

# Caracterizacion de la poblacion



## Genero de los Participantes


```python
G_count=df["SEXO"].value_counts()
```

```python
fig,(ax1,ax2)= plt.subplots(2,1,figsize=(15, 10))
fig.suptitle('Participantes BC Modalidad Familiar 2023 SEXO')
ax1.pie(G_count, autopct="%1.1f%%",pctdistance=1.5, labeldistance=1.2,
        )
ax1.legend(G_count.index, loc=10,bbox_to_anchor=(2,0.1))
ax2.set_title('Participantes BC Modalidad Familiar 2023 SEXO')
df['SEXO'].value_counts().sort_index().plot(kind='bar', color='blue', ax=ax2)
ax2.set_xlabel('Estado de los participantes')
ax2.set_ylabel('# Participantes')
```

```python
pd.crosstab(index=df['SEXO'], columns="count")
```
